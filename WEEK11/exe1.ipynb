```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier


data = pd.read_csv('Data.csv')


print("Original Dataset:")
print(data)


X = data[['Age', 'Income']]
y = data['Risk']


scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)


scaled_data = pd.DataFrame(X_scaled, columns=['Age', 'Income'])
print("\nMin-Max Scaled Dataset:")
print(scaled_data)


new_record = pd.DataFrame({'Age': [66], 'Income': [36120.34]})
new_record_scaled = scaler.transform(new_record)

#euclidean
knn_classifier = KNeighborsClassifier(n_neighbors=9,metric='euclidean')
knn_classifier.fit(X_scaled, y)


predicted_label = knn_classifier.predict(new_record_scaled)

print("\nEuclidean\n")

print("Predicted Risk for Record #10:", predicted_label[0])

#manhattan
knn_classifier = KNeighborsClassifier(n_neighbors=9,metric='manhattan')
knn_classifier.fit(X_scaled, y)


predicted_label = knn_classifier.predict(new_record_scaled)

print("\nManhattan\n")

print("Predicted Risk for Record #10:", predicted_label[0])

#chebyshev
knn_classifier = KNeighborsClassifier(n_neighbors=9,metric='chebyshev')
knn_classifier.fit(X_scaled, y)


predicted_label = knn_classifier.predict(new_record_scaled)

print("\nChebyshev\n")

print("Predicted Risk for Record #10:", predicted_label[0])

```
