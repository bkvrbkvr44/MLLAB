```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Wine dataset
wine_data = pd.read_csv("wine.csv")

# Standardize the data
wine_data_std = StandardScaler().fit_transform(wine_data)

# PCA
pca = PCA(n_components=2)
wine_data_pca = pca.fit_transform(wine_data_std)

# Print the explained variance ratio
print(pca.explained_variance_ratio_)

# Plot the data in the new principal component space
plt.scatter(wine_data_pca[:, 0], wine_data_pca[:, 1])
plt.xlabel("Principal component 1")
plt.ylabel("Principal component 2")
plt.title("Wine dataset in principal component space")
plt.show()

# Create a heatmap of the correlation matrix of the original data
correlation_matrix = wine_data.corr()
plt.imshow(correlation_matrix, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(wine_data.columns)), wine_data.columns, rotation=45)
plt.yticks(range(len(wine_data.columns)), wine_data.columns)
plt.title("Correlation matrix of Wine dataset")
plt.show()
```
