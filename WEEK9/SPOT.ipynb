```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
import seaborn as sns
```

```python
plt.style.use('seaborn')
data = pd.read_csv('housing.csv')
cols=['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_house_value','ocean_proximity']
data.drop(cols, axis=1, inplace=True)
data.head()
```

```python
clustering_data = data
from sklearn.cluster import KMeans
wcss=[]
```

```python
for i in range(1,11):
    km = KMeans(i)
    km.fit(clustering_data)
    wcss.append(km.inertia_)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax = plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.axvline(x=3, ls='--')
plt.ylabel('WCSS')
plt.xlabel('No. of Clusters (k)')
plt.title('The Elbow Method', fontsize = 20)
plt.show()
```

```python
from sklearn.cluster import KMeans
kms = KMeans(n_clusters=3, init='k-means++',max_iter=10)
kms.fit(clustering_data)
clusters = clustering_data.copy()
clusters['Cluster_Prediction'] = kms.fit_predict(clustering_data)
cluster_means = clusters.groupby('Cluster_Prediction').mean()
print(cluster_means)
```
