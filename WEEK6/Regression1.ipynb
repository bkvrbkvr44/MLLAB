```python
import matplotlib.pyplot as plt

# Data
Id = [0.734, 0.886, 1.04, 1.19, 1.35, 1.50, 1.66, 1.81, 1.97, 2.12]
Vgs = [1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0]

# Scatter plot
plt.scatter(Vgs, Id)
plt.xlabel('Gate-to-Source Voltage (Vgs)')
plt.ylabel('Drain Current (Id) (mA)')
plt.title('Scatter Diagram')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Reshape data for sklearn
Vgs = np.array(Vgs).reshape(-1, 1)
Id = np.array(Id)

# Create and fit the model
model = LinearRegression()
model.fit(Vgs, Id)

# Get the slope (m) and intercept (b)
m = model.coef_[0]
b = model.intercept_

print(f"The simple linear regression model is: Id = {m:.4f} * Vgs + {b:.4f}")
```
