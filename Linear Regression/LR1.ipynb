```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Create a Linear Regression model
lin_reg = LinearRegression()

# Fit the model to the data
lin_reg.fit(X, y)

# Make predictions using the model
X_new = np.array([[0], [2]])
y_predict = lin_reg.predict(X_new)

# Plot the data and the regression line
plt.scatter(X, y)
plt.plot(X_new, y_predict, 'r-')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
# Slope (Coefficient)
slope = lin_reg.coef_[0][0]

# Intercept
intercept = lin_reg.intercept_
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Calculate mean and variance of y
mean_y = np.mean(y)
variance_y = np.var(y)
print(f"\nMean of y: {mean_y}")
print(f"Variance of y: {variance_y}")

# Make predictions for the entire dataset
y_predictions_all = lin_reg.predict(X)
# Print the actual predictions made by the model
print("Actual Predictions:")
print(y_predictions_all)

# New value of x for prediction
new_x_value = np.array([[4]])
# Predict y for the new x value
predicted_y_new = lin_reg.predict(new_x_value)
print(f"Predicted y for x={new_x_value[0][0]}: {predicted_y_new[0][0]}")
```
